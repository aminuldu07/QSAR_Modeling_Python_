# Calculate Z scores 


This is the fourth notebook in a series of Jupyter Notebooks linked to the project titled "__Developing Mechanism-Based Models for Complex Toxicology Study Endpoints Using Standardized Electronic Submission Data__".  It only requires that the first notebook to create the training set be run.

This notebook will calculate Z scores for each descriptor relative to different disease liver phenotypes.  A Z score can be thought of as the mean of the disease positive animals relative to the total mean expressed in terms of standard deviations.  

The calculation comes from Kevin Cross of Leadscope and is defined as follows:

$$
z score = (\mu_{subset} - \mu_{fullset}) * \sqrt{(N_{subset} * N_{fullset}) / ({\sigma^{2}}_{fullset} * (N_{fullset} - N_{subset})}
$$


```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
import math, os
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats
```

```python
# Define species to make a training set
# and make a seperate folder to store
# all the resulting data

species = 'RAT'

if not os.path.exists('data'):
    os.mkdir('data')
    
species_data = os.path.join('data', species)
if not os.path.exists(species_data):
    os.mkdir(species_data)
    
training_data_file = os.path.join(species_data, f'{species}_training_data.csv')
```

Prepare the data as we do when modeling, except there is not need to standardize.

```python
min_response_value = 0.4

df = pd.read_csv(training_data_file, index_col=0)
df = df.replace(np.inf, np.nan)
srted_tests = df.notnull().sum().sort_values(ascending=False)

good_tests = df.columns[(df.notnull().sum() / df.shape[0]) > min_response_value]
good_tests = good_tests[~good_tests.isin(['USUBJID', 'STUDYID', 'SEX', 'STEATOSIS',
                                         'CHOLESTASIS', 'NECROSIS', 'SPECIES', 'IS_CONTROL',
                                         'BWDIFF', 'BWSLOPE', 'BWINTCEPT', 'MISTRESC'])]

data = df[good_tests]
data = data.apply(lambda x: x + abs(x.min()) + 1)
data = data.applymap(math.log10)


data.index = df.USUBJID
data['SEX'] = df['SEX']

le = LabelEncoder()
scaler = StandardScaler()


data['SEX'] = le.fit_transform(data['SEX'])

data.head()
```

```python
diseases = ['NECROSIS', 'CHOLESTASIS', 'STEATOSIS']

zscores = pd.DataFrame()
pvalues = pd.DataFrame()

# remove sex
data = data.drop('SEX', axis=1)

for d_name in diseases:

    disease = df[d_name]
    disease.index = df.USUBJID

    for test in data.columns:
        stat, pval = scipy.stats.ttest_ind(data.loc[disease == 0, test].dropna(), data.loc[disease == 1, test].dropna())
        subset_mean = data.loc[disease == 1, test].mean()
        fullset_mean = data[test].mean()
        subset_count = data.loc[disease == 1, test].dropna().shape[0]
        fullset_count = data[test].dropna().shape[0]
        fullset_var = data[test].var()

        zscores.loc[test, d_name] = (subset_mean - fullset_mean) * math.sqrt((subset_count * fullset_count) / (fullset_var * (fullset_count - subset_count)))
        pvalues.loc[test, d_name] = pval

```

```python
figure_folder = os.path.join(species_data, 'figures')
if not os.path.exists(figure_folder):
    os.mkdir(figure_folder)
```

```python
from matplotlib.colors import DivergingNorm

axid = 0

norm = DivergingNorm(vmin=zscores.min().min(), vcenter=0, vmax=zscores.max().max())
min_z, max_z = zscores.min().min(), zscores.max().max()
n_features = data.columns

for d_name in diseases:
    fig, axarr = plt.subplots(figsize=(10, 14))
    z_data = zscores[d_name]
    pval_data = pvalues[d_name]

    # sns.set_context(rc = {'patch.edgecolor': 'k'})

    colors = [plt.cm.seismic(norm(c)) for c in z_data.sort_values(ascending=False)]
    g = sns.barplot(y=z_data.sort_values(ascending=False).index, x= z_data.sort_values(ascending=False), palette=colors, ax=axarr)
    axarr.axvline(0, color="k", clip_on=False)
    plt.setp(axarr.patches, edgecolor='k')

    i = 0
    for name, val in z_data.sort_values(ascending=False).iteritems():

        if val < 0:
            ha = "left"
            x = 0.1
        else:
            ha = "right"
            x = -0.1
        
        g.text(x, i, name, color='black', ha=ha,
               va='center', fontsize=14, fontstyle='italic')


        if val < 0:
            ha = "right"
        else:
            ha = "left"

        if pval_data[name] < (0.001/len(n_features)):
            g.text(val, i, "***", color='black', ha=ha,
                   va='center', fontsize=16, fontstyle='italic')
        elif pval_data[name] < (0.01/len(n_features)):
            g.text(val, i, "**", color='black', ha=ha,
                   va='center', fontsize=16, fontstyle='italic')
        elif pval_data[name] < (0.05/len(n_features)):
            g.text(val, i, "*", color='black', ha=ha,
                   va='center', fontsize=16, fontstyle='italic')


        # if pval_data[name] < (0.05/len(n_features)):
        #     g.text(val, i, "*", color='black', ha=ha,
        #            va='center', fontsize=20, fontstyle='italic')

        i = i + 1

    axarr.set_yticks([])
    axarr.set_xlim(min_z-1.5, max_z+1.5)

    axarr.set_xticks([min_z, 0, max_z])
    axarr.set_xticklabels(['Min', 0, 'Max'], fontsize=16)

    # axarr.set_xticklabels(axarr.get_xticks(), fontsize=16)
    # axarr.set_xticklabels([], fontsize=16)
    axarr.set_xlabel("Z-Score", fontsize=18)
    # axarr.figure.colorbar(sm)
    plt.tight_layout()
    
    if not os.path.exists(os.path.join(figure_folder, d_name)):
        os.mkdir(os.path.join(figure_folder, d_name))
    plt.savefig(os.path.join(figure_folder, d_name, 'z_scores.png'), transparent=True)
```
